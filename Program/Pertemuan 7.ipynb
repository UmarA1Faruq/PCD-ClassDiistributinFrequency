```python
import skimage
from skimage import io, restoration, color
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Memuat gambar yang rusak
damaged_image_path = "/content/damaged_astronaut.png"  # Ganti dengan path gambar Anda
damaged_image = io.imread(damaged_image_path)
```

```python
# Menghapus kanal alpha jika ada (hanya menggunakan RGB)
damaged_image_rgb = damaged_image[..., :3]
```

```python
# Mengubah gambar menjadi grayscale
damaged_image_gray = color.rgb2gray(damaged_image_rgb)
```

```python
# Membuat mask di mana piksel hitam (bernilai 0) ditandai sebagai area rusak
mask = damaged_image_gray == 0
```

```python
# Menerapkan fungsi inpainting
restored_image = restoration.inpaint_biharmonic(damaged_image_gray, mask)
```

```python
# Menampilkan gambar asli, mask, dan gambar hasil restorasi untuk perbandingan
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(damaged_image_gray, cmap='gray')
axes[0].set_title("Damaged Image (Grayscale)")
axes[0].axis("off")

axes[1].imshow(mask, cmap='gray')
axes[1].set_title("Mask")
axes[1].axis("off")

axes[2].imshow(restored_image, cmap='gray')
axes[2].set_title("Restored Image")
axes[2].axis("off")

plt.show()
```

```python
from skimage import io, restoration
import numpy as np
import matplotlib.pyplot as plt

# Memuat gambar yang berisi logo
image_path = "/content/new-logo.png"  # Ganti dengan path gambar yang sesuai
image_with_logo = io.imread(image_path)

# Konversi ke grayscale
# gray_image_with_logo = color.rgb2gray(image_with_logo)

# Menginisialisasi mask dengan ukuran yang sama dengan gambar (tanpa kanal warna terakhir)
mask = np.zeros(image_with_logo.shape[:2], dtype=bool)

# Menetapkan area logo untuk mask (posisi mungkin perlu disesuaikan tergantung gambar)
mask[210:290, 360:425] = True

# Menerapkan inpainting untuk menghapus logo
image_logo_removed = restoration.inpaint_biharmonic(image_with_logo, mask, channel_axis=-1)

# Menampilkan gambar asli dan gambar hasil restorasi
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(image_with_logo, cmap = 'gray')
axes[0].set_title("Image with Logo")
axes[0].axis("off")

axes[1].imshow(image_logo_removed)
axes[1].set_title("Image with Logo Removed")
axes[1].axis("off")
plt.show()

# Tugas ini dikerjakan oleh 22312115 Umar Al Faruq
```

```python
from skimage.segmentation import slic
from skimage.color import label2rgb
import matplotlib.pyplot as plt
```

```python
# memmbaca Gambar
image = io.imread('/content/chinese.jpg')
```

```python
# Mengaplikasi SLIC dengan 400 Regions
segments = slic(image, n_segments=5000)
```

```python
segmented_image = label2rgb(segments, image, kind='avg')
```

```python
def plot(image, title=''):
  plt.imshow(image)
  plt.axis('off')
  plt.title(title)
  plt.show()
```

```python
plot(image, title='Gambar Asli')
plot(segmented_image, title='Gambar Hasil Segmentasi')
```
